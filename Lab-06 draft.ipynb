# Importing libraries

```python
#Classifying textual data using Na√Øve Bayes method
%matplotlib inline
from sklearn.datasets import load_iris #We are importing a dataset using load_iris function of sklearn. The following data set consists of 3 different type of flowers (labels) with 50 features each
from sklearn.model_selection import train_test_split #import the train_test_split function from sklearn that we will utilize to split the data into training and testing set
from sklearn.naive_bayes import GaussianNB #We have different types of naive bayes methods and we will be utilizing the GaussianNB type
from sklearn import metrics #We will utilize this to find the accuracy score for our Naive Bayes method
import numpy as np #Library to manage arrays. We create an object of numpy named np
import matplotlib.pyplot as plt  #Responsibile for plotting of graphs and visualizaton of data. We create an object of it named plt
import seaborn as sns; sns.set() #another library for plotting graphs, we use it for plotting confusion matrix, we used an abreviation sns for the library
from sklearn.datasets import make_blobs #we use this function to generate random data
from sklearn.datasets import fetch_20newsgroups #Import dataset which consists of 18846 news article and 20 different labels categorizing the articles
from sklearn.feature_extraction.text import TfidfVectorizer #As for naive bays doesnt have pre-available features, we have to find them ourselves. TfidfVectorizer will extract features from document and give it to a compiler in the form of vector
from sklearn.naive_bayes import MultinomialNB #We have different types of naive bayes methods and we will be utilizing the MultinomialNB type
from sklearn.pipeline import make_pipeline #function to construct a Pipeline from the given estimators
from sklearn.metrics import confusion_matrix #we use this function to create a confusion matrix
import pandas as pd #We will use it to load the comma seperated file
```

# Determine the best accuracy:

```python
iris = load_iris() 
  
X = iris.data 
y = iris.target 
  
for i in range(0,15): 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i) 
    
    gnb = GaussianNB() 
    gnb.fit(X_train, y_train) 

    y_pred = gnb.predict(X_test) 

    print('Gaussian Naive Bayes model accuracy using random state {}: {}%'.format(i, metrics.accuracy_score(y_test, y_pred)*100))
    
    


for i in range(5,55,5):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=i/100, random_state=1) 
    
    gnb.fit(X_train, y_train) 

    y_pred = gnb.predict(X_test) 
    print('Gaussian Naive Bayes model accuracy using {}% of data for testing: {}%'.format(i, metrics.accuracy_score(y_test, y_pred)*100))
```

```python
data = fetch_20newsgroups() 
data.target_names 
```

```python
categories = ['talk.religion.misc', 'soc.religion.christian',
              'sci.space', 'comp.graphics'] 
train = fetch_20newsgroups(subset='train', categories=categories) 
test = fetch_20newsgroups(subset='test', categories=categories)
```

```python
print(train.data[5]) 
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB()) 
```

```python
model.fit(train.data, train.target) 
labels = model.predict(test.data) 
```

```python
mat = confusion_matrix(test.target, labels) 


sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, 
            xticklabels=train.target_names, yticklabels=train.target_names) 
plt.xlabel('true label') 
plt.ylabel('predicted label'); 
plt.show()

sns.heatmap(mat, square=True, annot=True, fmt='d', cbar=False, 
            xticklabels=train.target_names, yticklabels=train.target_names) 
plt.ylabel('true label') 
plt.xlabel('predicted label'); 
plt.show()

sns.heatmap(mat.T, square=False, annot=True, fmt='d', cbar=False, 
            xticklabels=train.target_names, yticklabels=train.target_names) 
plt.xlabel('true label') 
plt.ylabel('predicted label'); 
plt.show()


sns.heatmap(mat.T, square=True, annot=False, fmt='d', cbar=False, 
            xticklabels=train.target_names, yticklabels=train.target_names) 
plt.xlabel('true label') 
plt.ylabel('predicted label'); 
plt.show()

sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=True, 
            xticklabels=train.target_names, yticklabels=train.target_names) 
plt.xlabel('true label') 
plt.ylabel('predicted label'); 
plt.show()

sns.heatmap(mat.T, square=True, annot=True, fmt='e', cbar=False, #adjusting fmt from d to e would display values as exponential instead of decimal
            xticklabels=train.target_names, yticklabels=train.target_names) 
plt.xlabel('true label') 
plt.ylabel('predicted label'); 
plt.show()
```

```python
def predict_category(s, train=train, model=model): 
    pred = model.predict([s]) 
    return train.target_names[pred[0]] 
```
